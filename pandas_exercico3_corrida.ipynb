```python
import pandas as pd
```

```python
dados = pd.read_json('/content/corrida_de_rua.json')
dados.head()
```

```python
dados = pd.json_normalize(dados['custos_evento_corrida_rua'])
dados
```

```python
colunas = list(dados.columns)
colunas
```

```python
dados = dados.explode(colunas[:])
dados
```

Resetar os indices.

```python
dados.reset_index(inplace = True,drop=True)
dados.head()
```

```python
dados.info()
```

```python
import numpy as np
```

```python
dados['quantidades']=dados['quantidades'].astype(np.int64)
```

```python
dados.info()
```

```python
dados['custo_estimado'] = dados['custo_estimado'].apply(lambda x: x.replace('$','').replace(',','').strip())
```

```python
dados['custo_estimado'] = dados['custo_estimado'].astype(np.float64)
```

```python
dados.info()
```

```python
dados.head()
```

```python
dados['descricao']
```

```python
dados['descricao'] = dados['descricao'].str.lower()
```

```python
dados.head()
```

```python
dados
```

```python
dados.to_json('corrida_de_rua_normalizado.json')
```
